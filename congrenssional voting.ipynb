```python
import pandas as pd

df=pd.read_csv('congressional_voting_dataset.csv')
df.head()
```

```python
updated_df=df.drop(columns='political_party',axis=1)
updated_df
```

```python
updated_df.info()
```

```python
updated_df.describe()
```

```python
total = updated_df.isnull().sum().sort_values(ascending=False)
total
```

```python
updated_df.replace('?','NaN',inplace=True)
updated_df
```

```python
col_mode=updated_df.mode()
col_mode
```

```python
for i in updated_df.columns:
    temp=col_mode[i][0]
    for j in updated_df.index:
        if updated_df.loc[j,i]=='NaN':
            updated_df.loc[j,i]=temp
updated_df
    
```

```python
target_set=set(df['political_party'])
target_set
```

```python
target_list=[]
targetcol_list=list(df['political_party'])
for i in targetcol_list:
    if i=='democrat':
        target_list.append(0)
    elif i=='republican':
        target_list.append(1)
target_list
```

```python
target_df=pd.DataFrame({'political_party':target_list})
target_df
```

```python
train_df=pd.DataFrame([])
for i in updated_df.columns:
    train_df = pd.concat([train_df,pd.get_dummies(updated_df[i], prefix=i)],axis=1)
train_df
```

# Training = 70 Testing = 30

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(train_df,target_df,test_size=0.3,random_state=42)
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
ypred=clf.predict(x_test)
acc=accuracy_score(y_test,ypred)
print('CART Accuracy =',acc)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Perceptron
from sklearn.naive_bayes import GaussianNB
```

```python

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(x_train, y_train)
y_predict= random_forest.predict(x_test)
randomforest_accuracy=accuracy_score(y_test, y_predict)
print("Random Forest Accuarcy = ",randomforest_accuracy)

logreg=LogisticRegression()
logreg.fit(x_train, y_train)
y_predict=logreg.predict(x_test)
logreg_accuracy=accuracy_score(y_test, y_predict)
print("Logistic Regression Accuarcy = ",logreg_accuracy)

perceptron=Perceptron(max_iter=5)
perceptron.fit(x_train, y_train)
y_predict=perceptron.predict(x_test)
perceptron_accuracy=accuracy_score(y_test, y_predict)
print("Perceptron Accuarcy = ",perceptron_accuracy)

gaussian=GaussianNB() 
gaussian.fit(x_train, y_train)  
y_predict=gaussian.predict(x_test)
guassian_accuracy=accuracy_score(y_test, y_predict)
print("Guassian Accuarcy = ",guassian_accuracy)
```
